## Import libraries

```python
from sklearn.datasets import fetch_openml
from scipy.ndimage.interpolation import shift
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Get data

```python
# Fetching MNIST Dataset
mnist = fetch_openml('mnist_784', version=1)

# Get the data and target
X, y = mnist["data"], mnist["target"]
```

## General shift function

```python
def shift_image(data, dy, dx):
    """
    Shifts the given image data by the specified displacement and displays the original and shifted images.

    Parameters:
        data (pandas.Series or numpy.ndarray): The image data, represented as a 1D array or Series.
        dy (int): The displacement value for the vertical (y) axis.
        dx (int): The displacement value for the horizontal (x) axis.

    Returns:
        None

    Displays:
        Two subplots showing the original image and the shifted image.
    """
    image = data.values.reshape(28, 28)  # Reshape the  image data to a 2D array
    shifted_image = shift(image, [dy, dx], cval=0)  # Perform image shifting operation

    # Plotting the images
    fig, ax = plt.subplots(2, 1)
    fig.subplots_adjust(hspace=0.5)  # Adjust vertical spacing between subplots
    ax[0].imshow(image, cmap=mpl.cm.binary, interpolation="nearest")
    ax[0].set_title('Original Image')
    ax[1].imshow(shifted_image, cmap=mpl.cm.binary, interpolation="nearest")
    ax[1].set_title('Shifted Image')
    plt.show()
```

## Shift image up

```python

# Shift image up: dy, dx = (-1, 0)

shift_image(X.iloc[4,:], -5, 0)
```

## Shift image down

```python
# Shift image down: dy, dx = (1, 0)

shift_image(X.iloc[10,:], 5,0)
```

## Shift image to the right

```python
# Shift image right: dy, dx = (0, 1)

shift_image(X.iloc[6,:], 0, 5)
```

## Shift image to the left

```python
# Shift image left: dy, dx = (0, -1)

shift_image(X.iloc[20,:], 0,-5)
```
